# Visualization: Scatterplot and Histogram

##### Scatterplots/Histograms: These can be used to visualize the distribution of engagement metrics (likes, replies, retweets, quote retweets) for tweets spreading misinformation/disinformation. What can be done is to plot the engagement metrics on the y-axis and time on the x-axis to observe any patterns or changes over time.

## Initializations

Import the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load the cleaned dataset into a pandas DataFrame:

```python
df = pd.read_csv("cleaned_dataset.csv")
```

Convert the "Date posted" column to datetime format:

```python
df["Date posted"] = pd.to_datetime(df["Date posted"])
```

## Plotting of Scatterplot

```python
# Scatterplot
plt.scatter(df["Date posted"], df["Likes"], label="Likes")
plt.scatter(df["Date posted"], df["Replies"], label="Replies")
plt.scatter(df["Date posted"], df["Retweets"], label="Retweets")
plt.scatter(df["Date posted"], df["Quote Tweets"], label="Quote Tweets")
plt.axvline(pd.Timestamp("2019-04-25"), color="red", linestyle="--", label="Fact-checking Articles")
plt.xlabel("Date posted")
plt.ylabel("Engagement Metrics")
plt.title("Engagement Metrics for Tweets over Time")
plt.legend()
plt.show()
```

##### In the scatterplot, the x-axis represents the "Date posted" (time) and the y-axis represents the respective engagement metric. Each engagement metric is plotted separately and displayed with different markers. The broken red line is the separator for when, the fact checking related articles related to the topic were released.

## Plotting of histogram

```python
# Histogram
plt.hist(df["Likes"], bins=20, alpha=0.5, label="Likes")
plt.hist(df["Replies"], bins=20, alpha=0.5, label="Replies")
plt.hist(df["Retweets"], bins=20, alpha=0.5, label="Retweets")
plt.hist(df["Quote Tweets"], bins=20, alpha=0.5, label="Quote Tweets")
plt.xlabel("Engagement Metrics")
plt.ylabel("Frequency")
plt.title("Distribution of Engagement Metrics")
plt.legend()
plt.show()
```

##### In the histogram, each engagement metric is plotted individually with bins representing the range of engagement values. The frequency of occurrence of each engagement metric within each bin is displayed.
